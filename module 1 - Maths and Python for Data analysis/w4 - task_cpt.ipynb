# Центральная предельная теорема для гамма-распределения

```python
# импорт необходимых библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

Плотность вероятности гамма-распределения задается формулой
$$\phi(x)=\frac{1}{\Gamma(\alpha)}x^{\alpha-1}e^{-x},$$
где $\alpha$ - положительный параметр, а $\Gamma(\alpha)$ - гамма-функция от $\alpha$. При этом и матожидание гамма-распредления и его дисперсия равны $\alpha$.

Создание выборки гамма-распределения:

```python
a = 5.0 # параметр гамма-распределения
rv = sts.gamma(a)

mu = a # матожидание 
sigma2 = a # дисперсия

sample = rv.rvs(1000)
```

Построение гистограммы и графика теоретической плотности распределения: 

```python
plt.hist(sample, bins=20, normed=True)

x = np.linspace(0,15,100)
pdf = rv.pdf(x)
plt.plot(x, pdf, linewidth=3.0)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Генерация 1000 выборок объема $n$ для $n=$5, 10 и 50:

```python
sample_sizes = np.array([5, 10, 50])
samples = {5:[], 10:[], 50:[]}
for n in sample_sizes:
    for i in xrange(1000):
        sample = rv.rvs(n)
        samples[n].append(sample)
```

Матожидание и дисперсия распределения заданы в переменных _mu_ и _sigma2_ соответственно. 
Согласно предельной теореме наши выборочные средние приближаются к следующим нормальным распределениям: 
$$\bar{X_{n}}\approx\sim N(EX,\frac{DX}{n})$$
где $EX=mu$ и $DX=sigma2$, а _n_ - объем выборки.

```python
# параметры нормальных распределений
sigmaz2=dict()
for n in sample_sizes:
    sigmaz2[n]=sigma2/n

print('для n=5: mu=%.2f, sigma2=%.2f' % (mu, sigmaz2[5]))
print('для n=10: mu=%.2f, sigma2=%.2f' % (mu, sigmaz2[10]))
print('для n=50: mu=%.2f, sigma2=%.2f' % (mu, sigmaz2[50]))

# создание соответствующих нормальных распределений
norm_rvz = dict()
for n in sample_sizes:
    norm_rvz[n] = sts.norm(loc=mu, scale=np.sqrt(sigmaz2[n]))
```

Построение гистограмм выборочных средних и плотностей соответствующих нормальных распределений:

```python
# вычисление выборочных средних
means = {5:[], 10:[], 50:[]}
for n in sample_sizes:
    for sample in samples[n]:
        m = np.mean(sample)
        means[n].append(m)


# построение гистограмм и графиков плотностей разбиения
for n in sample_sizes:
    print("n=%d:" % n)
    plt.hist(means[n], bins=20, normed=True)

    x = np.linspace(0,10,100)
    pdf = norm_rvz[n].pdf(x)
    plt.plot(x, pdf, linewidth=3.0)

    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    plt.show()
```

Как видно из графиков, при увеличении объема выборки $n$ распределение выборочных средних все лучше аппроксимируется соответствующим нормальным распределением. Таким образом мы убедились в том, что Центральная Предельная Теорема работает!
